Import libraries

```python
import pandas as pd
import numpy as np
```

Read in team wins file

```python
#change file format to CSV
team_wins = pd.read_csv(r'raw_files\win_totals\mlb_win_totals.csv')
```

Clean data file - this has over 90 years of data. I need just the last 5 years

```python
#drop rows to only include years 2019-2023
team_wins.drop(team_wins[team_wins['Year'] < 2019].index, inplace=True)

team_wins.head(10)
```

```python
#drop columns with no data (removing teams that do not exist any longer)
team_wins=team_wins.dropna(axis=1, how='all')

team_wins.head()
```

```python
#converting all values to interger (some are currenlty floats)
team_wins=team_wins.astype(int)
team_wins.drop([3, 4], inplace=True)

team_wins.head()
```

```python
#update CSV file
team_wins.to_csv(r'cleansed_data\mlb_win_totals.csv', index=None, header=True)
```
