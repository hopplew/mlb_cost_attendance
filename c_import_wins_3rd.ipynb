Import libraries that will be needed for this project

```python
import pandas as pd
import numpy as np
```

Read in team wins data file with Pandas
This file were built from data found at https://www.baseball-reference.com/leagues/majors/index.shtml and put into a CSV file 

```python
#read in team wins file
team_wins = pd.read_csv(r'raw_files\win_totals\mlb_win_totals.csv')
```

I can see this file has a lot of unneeded results as it goes back over 90 years, many of the teams on this list do not even exist any longer.
I just need the last 3 years of team's wins so it will need to be cleaned to hold only those results.

First step, as I know I only the last 3 years of data, I will drop any rows before 2021.

```python
#drop rows to only include years 2019-2023
team_wins.drop(team_wins[team_wins['Year'] < 2021].index, inplace=True)

team_wins.head()
```

Any team that has no data since 2021 can also be dropped as these would be historic teams and not relevant to this dataframe.

```python
#drop columns with no data (removing teams that do not exist any longer)
team_wins=team_wins.dropna(axis=1, how='all')

team_wins.head()
```

Next I can convert all the value types to interger for future data analysis

```python
#converting all values to interger (some are currenlty floats)
team_wins=team_wins.astype(int)

team_wins.head()
```

A relatively easy file to clean, this one is now ready to write to csv for use in building larger dataframe

```python
#update CSV file
team_wins.to_csv(r'cleansed_data\mlb_win_totals.csv', index=None, header=True)
```
